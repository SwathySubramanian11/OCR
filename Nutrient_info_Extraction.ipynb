```python
import easyocr

def recognize_text(img_path):
    '''loads an image and recognizes text.'''
    
    reader = easyocr.Reader(['en'], gpu=False)
    return reader.readtext(img_path)

# Perform OCR and get the results
result = recognize_text(im_1_path)

# Define the file path where you want to save the results
output_file_path = 'ocr_results.txt'  # Change the file extension to .csv for CSV format

# Open the file for writing
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    for bounding_box, text, confidence in result:
        # Write each result to the file in the desired format
        output_file.write(f"Bounding Box: {bounding_box}\n")
        output_file.write(f"Recognized Text: {text}\n")
        output_file.write(f"Confidence Score: {confidence}\n")
        output_file.write("\n")  # Add a separator between entries if needed
```

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# OCR results in the specified format
results = result

# Define custom entity patterns for nutrient names and quantities
ner_patterns = [
    {"label": "NUTRIENT", "pattern": [{"LOWER": {"in": ["calories", "fat", "saturated", "trans", "cholesterol", "sodium", "carbohydrate", "fiber", "sugars", "protein", "vitamin", "calcium", "iron", "potassium"]}}]},
    {"label": "QUANTITY", "pattern": [{"IS_DIGIT": True}, {"LOWER": "g"}]}
]

# Add custom NER patterns to the NLP pipeline
ruler = nlp.add_pipe("entity_ruler", before="ner")
ruler.add_patterns(ner_patterns)

# Iterate through OCR results and extract nutrient information
for bbox, recognized_text, confidence in results:
    # Preprocess the text (remove punctuation, convert to lowercase)
    preprocessed_text = recognized_text.lower()

    # Perform Named Entity Recognition (NER) using spaCy
    doc = nlp(preprocessed_text)

    # Initialize variables to track nutrient information
    current_nutrient = None
    nutrient_info = {}

    # Extract nutrient information using custom rules
    for ent in doc.ents:
        if ent.label_ == "NUTRIENT":
            current_nutrient = ent.text
        elif ent.label_ == "QUANTITY" and current_nutrient:
            nutrient_info[current_nutrient] = ent.text
            current_nutrient = None

    # Print the extracted nutrient information for this OCR result
   # print("Nutrient Information:")
    if nutrient_info:
        for nutrient, quantity in nutrient_info.items():
            print(f"{nutrient}: {quantity}")
```

```python
def read_nutrient_info(file_path):
    nutrient_info = {}
    with open(file_path, 'r') as file:
        for line in file:
            key, value = line.strip().split(': ')
            nutrient_info[key] = value
    return nutrient_info

file_path = './dataset/nutrient_info.txt'

# Debug statement to check the file path
print(f"Reading file from: {file_path}")

nutrient_info = read_nutrient_info(file_path)

# Debug statement to check if data was read
print("Nutrient Information:")
print(nutrient_info)

# Check if the dictionary is empty
if not nutrient_info:
    print("No nutrient information found in the file.")
else:
    # Print nutrient information
    for nutrient, value in nutrient_info.items():
        print(f"{nutrient}: {value}")
```

```python
nutrient_names = [
    "Calories", "Total Fat", "Saturated Fat", "Trans Fat",
    "Cholesterol", "Sodium", "Total Carbohydrate", "Dietary Fiber",
    "Total Sugars", "Added Sugars", "Protein", "Vitamin D",
    "Calcium", "Iron", "Potassium"
]

# Function to extract nutrient information from OCR result
def extract_nutrient_info(ocr_result):
    nutrient_info = {}
    recognized_text = '\n'.join([item[1] for item in ocr_result])
    
    for line in recognized_text.split('\n'):
        # Check if the line contains a nutrient name
        for nutrient in nutrient_names:
            if nutrient.lower() in line.lower():
                # Split the line to extract the value
                parts = line.split(': ')
                if len(parts) == 2:
                    nutrient_info[nutrient] = f"{nutrient}: {parts[1]}"
    
    return nutrient_info

# Example: Extract nutrient info from the provided OCR result
nutrient_info = extract_nutrient_info(result)

# Print or use the extracted nutrient information in the specified format
for nutrient, value in nutrient_info.items():
    print(value)
```
